# Morphology Analysis

Khaiii(Kakao Hangul Analyzer III) 형태소 분석기

```python
khai = KhaiiiApi()  # khaiii 형태소 분석기
```

```python
def make_corpus_show_morph_khaiii(text):
  corpus = []
  for s in text:
    sentence = []
    for word in khai.analyze(s):
      for morph in word.morphs:
        sentence.append(morph.lex+'/'+morph.tag)
    corpus.append(sentence)
  return corpus
```

```python
corpus_khaiii = make_corpus_show_morph_khaiii(df['speech'])
corpus_khaiii
```
